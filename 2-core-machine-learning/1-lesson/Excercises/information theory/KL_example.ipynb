```python
import numpy as np
```

```python
# use just float due to a deprecating warning
def KL(a, b):
  print(a)
  print(b)
  a = np.asarray(a, dtype=float)
  b = np.asarray(b, dtype=float)
  return np.sum(np.where(a != 0, a*np.log2(a/b),0))
```

```python
p = [9/25,12/25,4/25]
q = [1/3, 1/3, 1/3]
KL(p,q)
```
