```python
import pandas as pd
import numpy as np
```

```python

df = pd.DataFrame([[np.nan, 2, np.nan, 0],
                   [3, 4, np.nan, 1],
                   [np.nan, np.nan, np.nan, np.nan],
                   [np.nan, 3, np.nan, 4]],
                  columns=list("ABCD"))
```

```python
df
```

```python
df.fillna(0)
```

```python
df.fillna(method="ffill")
```

```python
values = {"A": 0, "B": 1, "C": 2, "D": 3}
```

```python
df.fillna(value=values)
```

```python
# Data encoding
```

```python
df["sex"] = ["M", "M", "F", "T"]
```

```python
df
```

```python
sex_map = {"M":1, "F":2, "T":3}
```

```python
df.sex = df.sex.apply(lambda x: sex_map[x])
```
