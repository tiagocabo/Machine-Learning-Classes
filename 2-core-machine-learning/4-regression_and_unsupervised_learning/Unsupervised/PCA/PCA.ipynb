```python
# lOAD IRIS DATASET
import pandas as pd
df = pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data",
                 names=['sepal length','sepal width','petal length','petal width','target'])
```

```python
df.info()
```

```python
df.describe()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
X = df[['sepal length','sepal width','petal length','petal width']]
y = df["target"]
X = sc.fit_transform(X)
```

```python
X
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
```

```python
import matplotlib.pyplot as plt
```

```python
X_reduced_components = pca.fit_transform(X)
```

```python
X_reduced_components
```

```python
principalDf = pd.DataFrame(data = X_reduced_components
             , columns = ['principal component 1', 'principal component 2'])
```

```python
principalDf
```

```python
targets = ['Iris-setosa', 'Iris-versicolor', 'Iris-virginica']
colors = ['r', 'g', 'b']
map_colors = dict(zip(targets, colors)) # {'Iris-setosa': 'r', 'Iris-versicolor': 'g', 'Iris-virginica': 'b'}
map_colors = y.apply(lambda x: map_colors[x])
```

```python
plt.scatter(principalDf['principal component 1'], principalDf['principal component 2'])
```

```python
pca.explained_variance_ratio_
```

```python
pca.components_
```

```python
import numpy as np
np.cumsum(pca.explained_variance_ratio_)
# Both PC explain 95% of the variance. This match the goods results we can see in the plot
```
