```python
from sklearn.cluster import KMeans
from sklearn import metrics
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Creating the data
x1 = np.array([3, 1, 1, 2, 1, 6, 6, 6, 5, 6, 7, 8, 9, 8, 9, 9, 8])
x2 = np.array([5, 4, 5, 6, 5, 8, 6, 7, 6, 7, 1, 2, 1, 2, 3, 2, 3])
X = np.array(list(zip(x1, x2))).reshape(len(x1), 2)
```

```python
# Visualizing the data
plt.plot()
plt.xlim([0, 10])
plt.ylim([0, 10])
plt.title('Dataset')
plt.scatter(x1, x2)
plt.show()
```

```python
inertias = []
mapping2 = {}
K = range(1, 10)
for k in K:
    # Building and fitting the model
    kmeanModel = KMeans(n_clusters=k).fit(X)

    inertias.append(kmeanModel.inertia_)

    mapping2[k] = kmeanModel.inertia_
```

```python

for key, val in mapping2.items():
    print(f'{key} : {val}')
```

```python
plt.plot(K, inertias, 'bx-')
plt.xlabel('Values of K')
plt.ylabel('Inertia')
plt.title('The Elbow Method using Inertia')
plt.show()
```

```python
# plot data with 3 clusters
kmeanModel = KMeans(n_clusters=3).fit(X)
```

```python
labels = kmeanModel.labels_
```

```python
# Visualizing the data
plt.plot()
plt.xlim([0, 10])
plt.ylim([0, 10])
plt.title('Dataset')
red_data_x1 = x1[labels == 0]
red_data_x2 = x2[labels == 0]

green_data_x1 = x1[labels == 1]
green_data_x2 = x2[labels == 1]

blue_data_x1 = x1[labels == 2]
blue_data_x2 = x2[labels == 2]

aa = plt.scatter(red_data_x1, red_data_x2, c="r", label="group 0")
aa = plt.scatter(green_data_x1, green_data_x2, c="g", label="group 1")
aa = plt.scatter(blue_data_x1, blue_data_x2, c="b", label="group 2")
plt.legend()
plt.show()
```
