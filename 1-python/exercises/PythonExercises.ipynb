# Beginner

1. Write a program that takes a user's name as input and prints out a greeting message.

2. Write a program that calculates the area of a rectangle given the length and width.



3. Write a program that takes a list of numbers as input and returns the sum of all the numbers in the list.



4. Write a program that takes a string as input and prints out the reverse of the string.



5. Write a program that takes a list of strings as input and returns the longest string in the list.



6. Write a program that takes two numbers as input and returns the larger of the two numbers.



7. Write a program that takes a list of numbers as input and returns the average of all the numbers in the list.

8. Write a program that takes a list of numbers as input and returns a new list with all the even numbers from the original list.

9. Write a program that takes a string as input and returns the number of vowels in the string.



10. Write a program that takes a list of strings as input and returns a new list with all the strings that have a length greater than 5.
